```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
dataset = pd.read_csv("asthamadataset.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
# dropping tables and renaming columns
dataset = dataset.drop(columns=['Severity_Mild', 'Severity_Moderate'])
dataset.rename(columns={'Severity_None': 'Target'}, inplace=True)
```

```python
# dataset = dataset.drop(columns=['None_Experiencing', 'None_Sympton'])
dataset
```

```python
x = dataset.drop(columns="Target")
x
y = dataset['Target']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, max_depth=None, random_state=42)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
