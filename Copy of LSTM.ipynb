```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
# load the dataset from google drive
# from google.colab import drive
# drive.mount('/content/drive')
# %cd /content/drive/MyDrive/final_project/E:/Vs Code/Django/projects/covid
data = pd.read_csv("E:/miniproject/final-disease-prediction/Diseaseprediction/Covid Dataset.csv")
boolean_columns = ['Breathing Problem', 'Fever', 'Dry Cough', 'Sore throat', 'Running Nose', 'Asthma',
                   'Chronic Lung Disease', 'Headache', 'Heart Disease', 'Diabetes', 'Hyper Tension', 'Fatigue', 'Gastrointestinal',
                    'Abroad travel', 'Contact with COVID Patient',
                   'Attended Large Gathering', 'Visited Public Exposed Places',
                   'Family working in Public Exposed Places', 'Wearing Masks', 'Sanitization from Market',
                   'COVID-19']

# Map "yes" to 1 and "no" to 0
data[boolean_columns] = data[boolean_columns].replace({'Yes': 1.0, 'No': 0.0})
```

```python
# split the data into features and target
features = data.drop(['Wearing Masks','Sanitization from Market','COVID-19'], axis=1)
features
```

```python
target = data["COVID-19"]
target
```

```python
from sklearn.model_selection import train_test_split
train_features, val_features, train_target, val_target = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
# pip install scikit-learn
```

```python
model = keras.Sequential()
from keras.layers import LSTM, Dense
model.add(keras.layers.Embedding(input_dim= 2, output_dim=128))
# first LSTM layer
model.add(LSTM(units=64, return_sequences=True))
# second LSTM layer
model.add(LSTM(units=32, return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#print(train_features)
print(train_target)
```

```python
history = model.fit(train_features, train_target, batch_size=64, epochs=50, validation_data=(val_features, val_target))
```

```python
# evaluate the model on the test data
val_loss, val_accuracy = model.evaluate(val_features, val_target)
print('Validation Loss: ', val_loss)
print('Validation Accuracy: ', val_accuracy)
```

```python
# %cd /content/drive/MyDrive/final_project/C:/Users/ASUS/Desktop/final_project
df = pd.read_csv("E:/miniproject/final-disease-prediction/Diseaseprediction/Covid Dataset.csv")
boolean_columns = ['Breathing Problem', 'Fever', 'Dry Cough', 'Sore throat', 'Running Nose', 'Asthma',
                   'Chronic Lung Disease', 'Headache', 'Heart Disease', 'Diabetes', 'Hyper Tension', 'Fatigue', 'Gastrointestinal',
                    'Abroad travel', 'Contact with COVID Patient',
                   'Attended Large Gathering', 'Visited Public Exposed Places',
                   'Family working in Public Exposed Places', 'Wearing Masks', 'Sanitization from Market',
                   'COVID-19']

# Map "yes" to 1 and "no" to 0
df[boolean_columns] = df[boolean_columns].replace({'Yes': 1.0, 'No': 0.0})
```

```python
test_features = df.drop(['Wearing Masks','Sanitization from Market' ,'COVID-19'], axis=1)
test_target = df["COVID-19"]
```

```python
test_loss, test_acc = model.evaluate(test_features, test_target)
print("test loss:" , test_loss) 
print('Test accuracy:', test_acc)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
test_pred = model.predict(test_features)
test_pred = (test_pred > 0.5).astype(int)
print(accuracy_score(test_target,test_pred))
```

```python
 #Precision
precision_positive = metrics.precision_score(test_target, test_pred, pos_label=1)
precision_negative = metrics.precision_score(test_target, test_pred, pos_label=0)
precision_positive, precision_negative
```

```python
#Recall
recall_sensitivity = metrics.recall_score(test_target, test_pred, pos_label=1)
recall_specificity = metrics.recall_score(test_target, test_pred, pos_label=0)
recall_sensitivity, recall_specificity 
```

```python
#F1 score
f1_positive = metrics.f1_score(test_target, test_pred, pos_label=1)
f1_negative = metrics.f1_score(test_target, test_pred, pos_label=0)
f1_positive, f1_negative
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_target, test_pred, target_names=['No', 'Yes']))
```
